```python
import pandas as pd
import plotly.express as px
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
#Reading the data as usual and cleaning it
```

```python
data = pd.read_csv('titanic.csv', sep=";")
```

```python
data.info()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.fillna(0, inplace=True)
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)

data = pd.get_dummies(data, columns=['Sex', 'Embarked'])

data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
#splitting data and training it
```

```python
x = data.drop("Survived",axis=1)
y = data["Survived"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#applying DecisionTree
```

```python
DT = DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
#visualizing the decision tree (plotting it)
```

```python
plt.figure(figsize=(15, 6))
plot_tree(DT, filled=True, feature_names=list(x.columns), class_names=['No', 'Yes'])
plt.title("Decision Tree")
plt.show()
```

```python
#changing the decisiontree parameteres 
```

```python
DT = DecisionTreeClassifier(max_depth=5, min_samples_split=5)
DT.fit(x_train, y_train)
```

```python
#calculating the accuracy
```

```python
y_pred = DT.predict(x_test)
accuracy_decision_tree = accuracy_score(y_test, y_pred)
print("Decision Tree Accuracy:", accuracy_decision_tree)
```

```python
#applying the randomforest methode
```

```python
RF = RandomForestClassifier(n_estimators=100)
RF.fit(x_train, y_train)
```

```python
#calculating again the accuracy
```

```python
y_pred = RF.predict(x_test)
accuracy_random_forest = accuracy_score(y_test, y_pred)
print("Random Forest Accuracy:", accuracy_random_forest)
```
