```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as pyo
import plotly.express as px
```

Q1 reading the csv file

```python
vg=pd.read_csv("Video games.csv")
```

```python
vg.dropna(inplace=True)
```

Q3 creating the area chart

```python
fig = px.area(vg, x='Year', y='Global_Sales', title='Trend of Global Sales Over the Years')

fig.update_xaxes(title_text='Year')
fig.update_yaxes(title_text='Global Sales')

fig.show()
```
