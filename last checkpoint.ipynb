```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
data = [
    ['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
    ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']
]
```

```python
te = TransactionEncoder()
te_trans = te.fit(data).transform(data)
df = pd.DataFrame(te_trans, columns=te.columns_)
```

```python
# Apply the apriori algorithm
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
```

```python
#association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
print(rules)
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv')
```

```python
dataset.fillna(0, inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
transactions = []
for i in range(len(dataset)):
    transactions.append([str(item) for item in dataset.iloc[i, :].dropna()])
```

```python
# Convert the dataset to a one-hot encoded format
te = TransactionEncoder()
te_ary = te.fit_transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Apply the apriori algorithm 
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
```

```python
# Association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
print(rules)
```
